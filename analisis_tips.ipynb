```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
print("\nFirst dataset rows: ")
display(df.head())
print("\nDataset Information: ")
print(df.info())
print("\nBasic Statistics")
display(df.describe)
```

# Informacion de la base de datos
## primer vistaso a los datos
La base de datos se conforma de una serie de categorias  como facturas,  propinas, sexo, fumadores, dias, hora, y tamaño.

# Filtrado
Hemos revisado si los datos requerian que se hiciera un filtrado, dado que valores en formatos no adecuados pueden dañar los analisis  subsecuentes

```python
sns.barplot(x="day", y="tip", data=df)
plt.title("Average day tip")
plt.figure(figsize=(6,4))
plt.show()
```

## Relación de propinas con el día de la semana
Cuando se quiere estudiar la realcion  de las propinas promedio respecto al dia de la semana, nos damos cuenta que hay una tendencia a que los domingos sean los dias donde mas propinas se obtienen, seguidopor el dia sabado. lo que tiene mas logica teniendo en cuenta que para estos dias los grupos de personas que suelen ir a comer tienden a ser mas grandes.


```python
sns.barplot(x="sex", y="tip", data=df)
plt.title("Average tip per sex")
plt.figure(figsize=(6,4))
plt.show()
```

## Relació de las propinas respecto a el genero.
Se puede observar que la variacion es muy ligera entre generos, sin embargo se puede observar qeu hay una ligera inclinacion a que los hombres dejen en promedio cerca de 20 centimos mas de propina respectoa  las mujeres

```python
sns.histplot(x="tip",bins=10, data=df)
plt.title("Histograma de datos")
plt.figure(figsize=(4,6))
plt.show()
```

## Comportamiento del tamaño de las propinas
Se puede ver que en realidad la tendencia es asimetrica a la izquierda lo que se traduce en que la mayor cantidad de propinas suelen ser de valores muy pequeños entre 1- 3 euros, y las propinas altas son escasas aunque no inexistentes. 

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex",  data=df)
plt.title("relacion entre cuenta total y propina")
plt.figure(figsize=(4,6))
plt.show()
          
          
          
```

## Relación entre cuenta total y las pripinas, separadas entre generos.
Se concluye que la tendencia es positiva por lo que cuando las cuentas son mucho mas grandes por consiguiente las propinas tambien, adicioanlmente s epuede observar que apesar de que la tendencia es positiva, la dispercion de los datos es grande, por lo que  hay una gran cantidad de casos que se salen de la media. tambien se puede observar que en los casos donde las mujeres pagaron la cuenta estas tendian a estar en un rango de valor total entre 1y 40 euros, siendo su grueso entre 10 y 30 euros.  caso simular que ocurre con los gombres  y algunos casos atipicos de facturas muy altas que dejaron por consiguiente propinas altas. 

```python
sns.boxplot(x="size", y="tip", data=df)
plt.title("propinas segun el tamaño del grupo")
plt.figure(figsize=(4,6))
plt.show()
```

# Análisis de propinas en restaurante

## Conclusiones principales
- **Distribución:** la mayoría de las propinas están entre 2 y 3, pocas son mayores de 6.
- **Día de la semana:** los domingos se dejan más propinas que entre semana.
- **Género:** los hombres dejan en promedio un poco más de propina que las mujeres.
- **Cuenta total:** existe una correlación positiva: a mayor cuenta, mayor propina.
- **Tamaño del grupo:** los grupos grandes dejan más propinas, aunque con mayor variabilidad.
